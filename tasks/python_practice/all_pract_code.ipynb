```python
from re import search

s = input("Enter the passwords : ").split(",")

lst =[]

for i in s:
    count = 0
    count += (len(i)>=6 and len(i)<=12)
    count += bool(search("[A-Z]",i))
    count += bool(search("[a-z]",i))
    count += bool(search("[0-9]",i))
    count += bool(search("[@$#]",i))
    
    if count==5:
        lst.append(i)
        
print(",".join(lst))
```

```python
from math import sqrt
vert = 0
hor = 0
while True:
    lst = input().split(' ')
    if not lst[0]:
        break
    if( lst[0]== "up"):
        vert += int (lst [1])
    elif (lst[0] == "down"):
        vert -= int (lst[1])
        
    elif(lst[0] == "right"):
        hor += int(lst[1])
    elif(lst[0] == "left"):
        hor -= int (lst[1])
        
total = int(sqrt((vert**2) + (hor**2)))
print (total)
```
