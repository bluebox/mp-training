```python
class mobile:
    def __init__(self,name,model_number,price,stock):
        self.name=name
        self.model_number=model_number
        self.price=price
        self.stock=stock
        
    def show(self):
        print(self.name ,'\t',self.model_number,"\t",self.price,"\t",self.stock)
        
        
p1 = mobile("iphone",13,75000,10)
p2 = mobile("realme",9,20000,50)
p3 = mobile("redmi",9,17000,45)
p=[p1,p2,p3]

class store:
    def __init__(self):
        pass
    
    def show_inventory(self):
        print("Name \t model \t price \t stock ")
        p1.show()
        p2.show()
        p3.show()
    
    def search(self):
        new=input("enter the brand name")
        for i in range(len(p)):
            if p[i].name==new:
                print("in stock")
                break
            else:
                print("out of stock")
    
    def update(self):
        new=input("enter the brand name")
        up=int(input("enter the stock update"))
        for i in range(len(p)):
            if p[i].name==new:
                p[i].stock=up
                break
            else:
                print("create a new object")
                break
    
    def delete(self):
        new=input("enter the brand name")
        for i in range(len(p)):
            if p[i].name==new:
                p.remove(new)
                break 
        
         
m_store=store()           
while True:
    print("press 1 for viewing the inventory")
    print("press 2 for searching the inventory")
    print("press 3 for updating the stock")
    print("press 4 for deleting the product")
    print("press 5 for exit")
    choice=int(input())
    if choice==1:
        m_store.show_inventory()
    elif choice==2:
        m_store.search()
    elif choice==3:
        m_store.update()
    elif choice==4:
        m_store.delete()
    elif choice==5:
        break
    else:
        print("invalid choice")
        continue
        
            
        

```
