```python
def p_check(password):
    up=low=num=special=0
    if(len(password)>=6 and len(password)<=12):
        for i in range(len(password)):
            if(password[i].isupper()):
                up+=1
            elif(password[i].islower()):
                low+=1
            elif(password[i].isdigit()):
                num+=1
            elif(password[i]=='$' or password[i]=='#' or password[i]=='@'):
                special+=1
    if(up>0 and low>0 and special>0 and num>0):
        return True
    else:
        return False
lst=input().split(",")        
            
new_list=[]
for i in lst:
    if(p_check(i)):
        new_list.append(i)
print(",".join(new_list))        
```

```python
li=[1, 2, 3, 4, 5, 6, 7, 8]

def binary(li, x):
    low, high=0, len(li)-1
    while(low<=high):
        mid=int(low+(high-low)/2)
        if(li[mid]==x):
            return mid
        elif(x<li[mid]):
            high=mid-1
        else:
            low=mid+1     
print(binary(li, 6))
```
